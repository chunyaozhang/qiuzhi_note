```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy import exp
from sklearn import datasets

#处理数据集iris
dataX=datasets.load_iris()['data'][0:100].tolist()
dataY=datasets.load_iris()['target'][0:100].tolist()
for i in range(len(dataX)):
    dataX=datasets.load_iris()['data'][0:100].tolist()
    dataY=datasets.load_iris()['target'][0:100].tolist()
for i in range(len(dataX)):
    dataX[i].insert(0,1)
    dataX[i].insert(5,dataY[i])
    
    
def plotLog(data,weights,ttype):
    x0=[]
    x1=[]
    y0=[]
    y1=[]
    x2=[]
    y2=[]
    for i in range(len(data)):
        data0=data[i]
        if(ttype[i]==False):
            x2.append(data0[1])
            y2.append(data0[2])
        elif(data0[5]==0):
            x0.append(data0[1])
            y0.append(data0[2])
        else:
            x1.append(data0[1])
            y1.append(data0[2])
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(x2,y2,marker='x',label='1',s=40,c='grey') #分错的点
    ax.scatter(x0,y0,marker='+',label='1',s=40,c='r')
    ax.scatter(x1,y1,marker='*',label='1',s=40,c='b')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.savefig('test.png')
    plt.show()





# sigmoid函数
def sigmoid(InX):
        return 1.0/(1+exp(-InX))
        


def Gradient_Descent(data,rate,iterNum):
    """
    data:list
    rate:学习率
    iterNum:迭代次数
    """
    x=[]
    y=[]
    weightList=[]
    err=[]
    #将自变量和因变量分别放在x,y中
    for i in range(len(data)):
        x.append(data[i][:-1])
        y.append(data[i][-1:])
    x=np.mat(x)
    y=np.mat(y)
    x2=x.T
    #给参数赋予初始权重
    weights=np.mat([[1],[1],[1],[1],[1]])
    for i in range(iterNum):
        h = sigmoid(x* weights)
        error=y-h
        weights=weights+rate*x2*error
        """
        以上代码迭代完毕
        ---------------
        下列代码存放结果
        """
        #存放权重
        wt=[]
        for s in range(len(weights)): 
            wt.append(weights[s,0])
        weightList.append(wt)
        #存放误差
        err.append(sum(error)[0,0])
        #画图
        ttype=[]
        for s in range(len(data)):
            if h[s,0]>0.5:
                m=1
            else:
                m=0
            a=(y[s,0]==m)
            ttype.append(a)
        plotLog(data,weights,ttype=ttype)
        #打印每次迭代每个样本y值，y估计值以及error
        #for j in range(len(y)):
            #print("第%s次迭代,第%s个样本\n{真实y[%s]:%s;估计y[%s]^:%s;error[%s]:%s;}"%(i.__str__(),j.__str__(),j.__str__(),y[j,0].__str__(),j.__str__(),h[j,0].__str__(),j.__str__(),error[j,0].__str__()))      
    return {'weights':weights,'weightList':weightList,'error':err,'y':y,'h':h}


    
```

```python
#预估结果与真实结果比对
result=Gradient_Descent(data=dataX,rate=0.05,iterNum=50)
```
