```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn import datasets
import math
```

```python
#step1:初始化中心点
def initCentr(data,k):
    """
    data: array
    k: 聚类的个数
    """
    n_Sample,dim=data.shape
    randNum=[random.randint(0,n_Sample) for i in range(k)] 
    return data[randNum,:]  #随机选择k个点
```

```python
#step2:计算中心点到每个点的距离，每个点选择离自己最近的点作为所属类
#定义两点间距离
def EUC_Dist(v1,v2):
    dist=math.sqrt(sum(pow(v1-v2,2)))  # sqrt(sum((v1-v2)*(v1-v2).T))
    return dist
```

```python
#更新聚类中心点
def update_Centr(data,k,cluster):
    centroid=np.zeros((k,data.shape[1]))
    for i in range(k):
        centroid[i,]=[data[[x==i for x in cluster],y].mean()  for y in range(data.shape[1])] 
    return centroid


def KMeans(data,k,iterNum):
    SSE={}
    for Niter in range(iterNum):
        if Niter==0:
            centroid=initCentr(data,k) #第一次迭代初始化中心点
        else:
            centroid=update_Centr(data=data,k=k,cluster=cluster)
            #计算每个点到中心的距离,打上类别标签
        cluster={}
        se=0
        for i in range(data.shape[0]):
            minDist=100000
            for j in range(k):
                dist=EUC_Dist(data[i,],centroid[j,]) # 样本点到各个中心点欧式距离
                if dist<minDist:
                    minDist=dist
                    cls=j
            se=se+minDist**2 # 计算每个样本到所属中心的距离
            cluster[i]=cls  # 打上类别标签
        cluster=pd.DataFrame.from_dict(cluster, orient='index').reset_index(drop=False)[0].tolist()
        SSE[Niter]=se
    centers=update_Centr(data=data,k=k,cluster=cluster)
    return {'labels_':cluster,'cluster_centers_':centers,'SSE_':SSE}
```

```python
#加载数据
iris=datasets.load_iris()
irisArray=iris.data
model=KMeans(data=irisArray,k=3,iterNum=100)
print(model['cluster_centers_'])
```

```python
#画plt.scatter(irisArray[:,0],irisArray[:,1])图
labels=model['labels_']
color={0:'pink',1:'blue',2:'yellow'}
colors=[color[i] for i in labels]
plt.scatter(irisArray[:,0],irisArray[:,1],color=colors,s=irisArray[:,2]*30,alpha=0.2)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.scatter(model['cluster_centers_'][0,0],model['cluster_centers_'][0,1],s=150,color='red',marker='+')
plt.scatter(model['cluster_centers_'][1,0],model['cluster_centers_'][1,1],s=150,color='blue',marker='+')
plt.scatter(model['cluster_centers_'][2,0],model['cluster_centers_'][2,1],s=150,color='brown',marker='+')
```
