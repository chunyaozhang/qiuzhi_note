# Age processing

```python
plt.violinplot(bank_df['AGE'], showmeans=True)
```

```python
plt.boxplot(bank_df['AGE'])
```

```python
for dataset in [bank_df]:
    dataset.loc[dataset['AGE'] <= 10, 'AGE'] = 0
    dataset.loc[(dataset['AGE'] > 10) & (dataset['AGE'] <= 20), 'AGE'] = 1
    dataset.loc[(dataset['AGE'] > 20) & (dataset['AGE'] <= 30), 'AGE'] = 2
    dataset.loc[(dataset['AGE'] > 30) & (dataset['AGE'] <= 40), 'AGE'] = 3
    dataset.loc[(dataset['AGE'] > 30) & (dataset['AGE'] <= 40), 'AGE'] = 4
    dataset.loc[(dataset['AGE'] > 40) & (dataset['AGE'] <= 50), 'AGE'] = 5
    dataset.loc[(dataset['AGE'] > 50) & (dataset['AGE'] <= 60), 'AGE'] = 6
    dataset.loc[(dataset['AGE'] > 60) & (dataset['AGE'] <= 70), 'AGE'] = 7
    dataset.loc[(dataset['AGE'] > 70) & (dataset['AGE'] <= 80), 'AGE'] = 8
    dataset.loc[(dataset['AGE'] > 80) & (dataset['AGE'] <= 90), 'AGE'] = 9
    dataset.loc[dataset['AGE'] > 90, 'AGE'] = 10
```

```python
df_age = pd.get_dummies(bank_df.AGE, prefix = 'AGE')
```

```python
bank_df_1 = bank_df.drop(['AGE'], axis=1)
```

```python
bank_df_2 = pd.concat((bank_df_1, df_age), axis=1)
```

# Modle preparation

```python
bankData_negtive=bank_df_9[bank_df_9.CHURN_CUST_IND==1]
bankData_postive=bank_df_9[bank_df_9.CHURN_CUST_IND==0].sample(bankData_negtive.__len__(),replace=False)
bankData_dealwith=pd.concat([bankData_negtive,bankData_postive])
```

```python
df_Y = bankData_dealwith['CHURN_CUST_IND']
```

```python
df_X = bankData_dealwith.drop(['CHURN_CUST_IND'], axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_X, df_Y, test_size=0.8, random_state=1)
```

```python
X_train.drop(['TELEBANK_ALL_TX_NUM','FR_SAV_CUR_ALL_BAL'], axis=1, inplace=True)
```

```python
X_test.drop(['TELEBANK_ALL_TX_NUM','FR_SAV_CUR_ALL_BAL'], axis=1, inplace=True)
```

# Model Processing

```python
pram_list = {
    'max_iter':list(range(100,100000,100)),
    'C':[0.01,0.1,1,10],
}
```

```python
#get best parameters
LogistcModel = GridSearchCV(LogisticRegression(),pram_list,cv=5)
LogistcModel.fit(X_train,Y_train)
print(LogistcModel.best_params_)
```

```python
print("....<1>First logistic regression model...")
LogistcModel_1=LogisticRegression(C=0.1, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=None, solver='lbfgs', tol=0.00001, verbose=0,
                   warm_start=False)
LogistcModel_1.fit(X_train,Y_train)
print(LogistcModel.score(X_train,Y_train))
```

```python
Y_pred = LogistcModel_1.predict(X_test)
acc_log = round(LogistcModel.score(X_train, Y_train)*100, 2)
acc_log
```

```python
Y_test.value_counts()
```

```python
print("....<2>Model prediction...")
print(pd.DataFrame(Y_pred)[0].value_counts())
print(Y_test.value_counts())
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
```

```python
fpr,tpr,threshold = metrics.roc_curve(Y_test, Y_pred) ###recall and precision
auc = metrics.auc(fpr,tpr)
```

```python
lw = 2
plt.figure(figsize=(10,10))
plt.plot(fpr, tpr, color='orange',lw=lw,label='ROC curve (area = %0.20f)' % auc) 
plt.plot([0, 1], [0, 1], color='lightgrey', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
